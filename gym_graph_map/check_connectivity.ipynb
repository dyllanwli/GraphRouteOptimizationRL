```python
import gym
import gym_graph_map

def avg_degree(edges):
    total_deg = sum([len(v) for v in edges])
    return total_deg / len(edges)



n_exp = 30

n = 1000
ks = range(1,10)
freqs = []
degs = []
def run():
    for k in ks:
        n_yes = 0
        total_deg = 0
        for _ in range(n_exp):
            g = gym.make("graph-search-rd-v0", n=n, k=k, d=2)
            if g.is_connected():
                n_yes += 1
            total_deg += avg_degree(g.graph_edges)
        freqs.append(n_yes/n_exp)
        degs.append(total_deg/n_exp)
run()
```
